```
#https://www.hackerrank.com/challenges/nested-list/submissions/code/457293739
students = []

n = int(input())

for _ in range(n):
    name = input()
    grade = float(input())
    students.append([name, grade])

grades = sorted(set([s[1] for s in students]))
second_lowest = grades[1]

names = [s[0] for s in students if s[1] == second_lowest]
names.sort()

for name in names:
    print(name)
################################################################################
#https://www.hackerrank.com/challenges/finding-the-percentage/submissions/code/457293334
n = int(input())

students = {}

for _ in range(n):
    data = input().split()
    name = data[0]
    marks = list(map(float, data[1:]))
    students[name] = marks

query_name = input()

avg = sum(students[query_name]) / len(students[query_name])
print(f"{avg:.2f}")

################################################################################
#https://www.hackerrank.com/challenges/nested-list/submissions/code/456890792
import sys

lines = [line.strip() for line in sys.stdin if line.strip()]

n = int(lines[0])
students = []

for i in range(1, 2*n, 2):
    name = lines[i]
    score = float(lines[i+1])
    students.append([name, score])

second_lowest = sorted(set(score for name, score in students))[1]
names = sorted(name for name, score in students if score == second_lowest)

for name in names:
    print(name)
################################################################################
#https://www.hackerrank.com/challenges/find-second-maximum-number-in-a-list/submissions/code/456889082
if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))

    unique_scores = list(set(arr))
    unique_scores.sort()
    print(unique_scores[-2])
################################################################################
#https://www.hackerrank.com/challenges/list-comprehensions/submissions/code/456888629
if __name__ == '__main__':
    x = int(input())
    y = int(input())
    z = int(input())
    n = int(input())
print([[i, j, k]
           for i in range(x + 1)
           for j in range(y + 1)
           for k in range(z + 1)
           if i + j + k != n])
################################################################################
#https://www.hackerrank.com/challenges/py-hello-world/submissions/code/456887789
print("Hello, World!")

################################################################################
#https://www.hackerrank.com/challenges/python-print/submissions/code/456886603
if __name__ == '__main__':
    n = int(input())
    for i in range (1,n +1 ):
        print (i,end="")
################################################################################
#https://www.hackerrank.com/challenges/validating-credit-card-number/submissions/code/456580760
# Enter your code here. Read input from STDIN. Print output to STDOUTimport re
import re
def validate_credit_card(card):

    pattern = r'^[456]\d{15}$|^[456]\d{3}(-\d{4}){3}$'
    if not re.match(pattern, card):
        return "Invalid"

    digits_only = card.replace('-', '')


    if re.search(r'(\d)\1{3,}', digits_only):
        return "Invalid"

    return "Valid"

n = int(input())
for _ in range(n):
    card = input().strip()
    print(validate_credit_card(card))
################################################################################
#https://www.hackerrank.com/challenges/validating-credit-card-number/submissions/code/456580757
# Enter your code here. Read input from STDIN. Print output to STDOUTimport re
import re
def validate_credit_card(card):

    pattern = r'^[456]\d{15}$|^[456]\d{3}(-\d{4}){3}$'
    if not re.match(pattern, card):
        return "Invalid"

    digits_only = card.replace('-', '')


    if re.search(r'(\d)\1{3,}', digits_only):
        return "Invalid"

    return "Valid"

n = int(input())
for _ in range(n):
    card = input().strip()
    print(validate_credit_card(card))
################################################################################
#https://www.hackerrank.com/challenges/validating-uid/submissions/code/456580430
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

def validate_uid(uid):

    if len(uid) != 10:
        return "Invalid"

    if len(set(uid)) != 10:
        return "Invalid"

    if len(re.findall(r'[A-Z]', uid)) < 2:
        return "Invalid"

    if len(re.findall(r'\d', uid)) < 3:
        return "Invalid"

    if not re.fullmatch(r'[A-Za-z0-9]+', uid):
        return "Invalid"

    return "Valid"
n = int(input())
for _ in range(n):
    uid = input().strip()
    print(validate_uid(uid))
################################################################################
#https://www.hackerrank.com/challenges/detect-html-tags-attributes-and-attribute-values/submissions/code/456580377
# Enter your code here. Read input from STDIN. Print output to STDOUT
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print(tag)
        for attr, value in attrs:
            print(f"-> {attr} > {value}")

    def handle_startendtag(self, tag, attrs):

        print(tag)
        for attr, value in attrs:
            print(f"-> {attr} > {value}")


html = ""
for _ in range(int(input())):
    html += input().rstrip() + "\n"

parser = MyHTMLParser()
parser.feed(html)
parser.close()
################################################################################
#https://www.hackerrank.com/challenges/html-parser-part-2/submissions/code/456580339
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        if '\n' in data:
            print(">>> Multi-line Comment")
            for line in data.splitlines():
                print(line)
        else:
            print(">>> Single-line Comment")
            print(data)

    def handle_data(self, data):
        if data != '\n':
            print(">>> Data")
            print(data)


html = ""
for _ in range(int(input())):
    html += input().rstrip()
    html += '\n'

parser = MyHTMLParser()
parser.feed(html)
parser.close()
################################################################################
#https://www.hackerrank.com/challenges/html-parser-part-1/submissions/code/456580216
# Enter your code here. Read input from STDIN. Print output to STDOUT
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):

    def handle_starttag(self, tag, attrs):
        print(f"Start : {tag}")
        for attr in attrs:
            name, value = attr
            print(f"-> {name} > {value}")

    def handle_endtag(self, tag):
        print(f"End   : {tag}")

    def handle_startendtag(self, tag, attrs):
        print(f"Empty : {tag}")
        for attr in attrs:
            name, value = attr
            print(f"-> {name} > {value}")


n = int(input())
html_lines = ""
for _ in range(n):
    html_lines += input()

parser = MyHTMLParser()
parser.feed(html_lines)
###############################################################################
#https://www.hackerrank.com/challenges/hex-color-code/submissions/code/456580172
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

n = int(input())
lines = [input() for _ in range(n)]

for line in lines:

    matches = re.findall(r'(?<=[:\s])#[0-9a-fA-F]{3,6}', line)
    for color in matches:
        print(color)
################################################################################
#https://www.hackerrank.com/challenges/validating-named-email-addresses/submissions/code/456580134
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re
import email.utils

n = int(input())
for _ in range(n):
    line = input()

    name, addr = email.utils.parseaddr(line)

    if re.match(r'^[a-zA-Z][\w\.-]*@[a-zA-Z]+\.[a-zA-Z]{1,3}$', addr):
        print(line)
################################################################################
#https://www.hackerrank.com/challenges/validating-the-phone-number/submissions/code/456580104
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

n = int(input())
for _ in range(n):
    s = input()
    if re.match(r'^[987][0-9]{9}$', s):
        print("YES")
    else:
        print("NO")
################################################################################
#https://www.hackerrank.com/challenges/validate-a-roman-number/submissions/code/456580081
regex_pattern = r"^M{0,3}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$"    # Do not delete 'r'.
################################################################################
#https://www.hackerrank.com/challenges/re-sub-regex-substitution/submissions/code/456579997
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

n = int(input())
lines = [input() for _ in range(n)]

for line in lines:
    line = re.sub(r'(?<= )&&(?= )', 'and', line)
    line = re.sub(r'(?<= )\|\|(?= )', 'or', line)
    print(line)
################################################################################
#https://www.hackerrank.com/challenges/re-start-re-end/submissions/code/456579933
import re

S = input()
k = input()

matches = list(re.finditer(r'(?={})'.format(re.escape(k)), S))

if matches:
    for m in matches:
        start = m.start()
        end = start + len(k) - 1
        print((start, end))
else:
    print((-1, -1))
################################################################################
#https://www.hackerrank.com/challenges/re-findall-re-finditer/submissions/code/456579518
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

s = input()

pattern = r'(?<=[b-df-hj-np-tv-zB-DF-HJ-NP-TV-Z])([aeiouAEIOU]{2,})(?=[b-df-hj-np-tv-zB-DF-HJ-NP-TV-Z])'

matches = re.findall(pattern, s)

if matches:
    for m in matches:
        print(m)
else:
    print(-1)
################################################################################
#https://www.hackerrank.com/challenges/re-group-groups/submissions/code/456579481
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re


s = input()

pattern = r'([a-zA-Z0-9])\1'

match = re.search(pattern, s)

if match:
    print(match.group(1))
else:
    print(-1)
################################################################################
#https://www.hackerrank.com/challenges/re-split/submissions/code/456579305
regex_pattern = r"[.,]"	# Do not delete 'r'.
################################################################################
#https://www.hackerrank.com/challenges/python-print/submissions/code/456579202
if __name__ == '__main__':
    n = int(input())
    for i in range (1,n +1 ):
        print (i,end="")
################################################################################
#https://www.hackerrank.com/challenges/write-a-function/submissions/code/456579054
def is_leap(year):
    if year % 4 != 0:
        return False
    elif year % 100 != 0:
        return True
    elif year % 400 == 0:
        return True
    else:
        return False
################################################################################
#https://www.hackerrank.com/challenges/python-loops/submissions/code/456578763
if __name__ == '__main__':
    n = int(input())
    for i in range(n) :
        print (i**2)
################################################################################
#https://www.hackerrank.com/challenges/python-division/submissions/code/456578679
if __name__ == '__main__':
    a = int(input())
    b = int(input())
    print(a//b)
    print(a/b)
################################################################################
#https://www.hackerrank.com/challenges/python-arithmetic-operators/submissions/code/456578633
if __name__ == '__main__':
    a = int(input())
    b = int(input())
    print(a+b)
    print(a-b)
    print(a*b)
################################################################################
#https://www.hackerrank.com/challenges/py-if-else/submissions/code/456578524
n = int(input())

if n % 2 != 0:
    print("Weird")
else:
    if 2 <= n <= 5:
        print("Not Weird")
    elif 6 <= n <= 20:
        print("Weird")
    elif n > 20:
        print("Not Weird")
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%#
```
