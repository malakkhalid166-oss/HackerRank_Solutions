```
#3-Prime number in Range
start = 1
end = 20
for num in range(start, end+1):
    if num < 2:
        continue
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
print()
```

```
#4
n = 10
fib = [0, 1]

for i in range(2, n):
    next_value = fib[i-1] + fib[i-2]
    fib.append(next_value)

print("Fibonacci sequence:", fib)
```

```
#8
while True:
    num = input("Enter a positive number: ")
    valid = True
    for char in num:
        if char not in "0123456789":
            valid = False
            break
    if valid and int(num) > 0:
        print("Valid input:", num)
        break
    else:
        print("Invalid input, try again.")
```

```
#11
num = int(input("enter number to check if it is palindrom or no:"))
original_num = num
reversed_num = 0

while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num = num // 10

if original_num == reversed_num:
    print(original_num, "is a palindrome")
else:
    print(original_num, "is not a palindrome")
```
