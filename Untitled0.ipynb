```python
#Lists
if __name__ == '__main__':
    N = int(input())
    my_list = []

    for _ in range(N):
        command_line = input().split()
        command = command_line[0]
        args = command_line[1:]
        args = list(map(int, args))
        if command == "insert":
            my_list.insert(args[0], args[1])
        elif command == "print":
            print(my_list)
        elif command == "remove":
            my_list.remove(args[0])
        elif command == "append":
            my_list.append(args[0])
        elif command == "sort":
            my_list.sort()
        elif command == "pop":
            my_list.pop()
        elif command == "reverse":
            my_list.reverse()
```

```python
#Nested Lists
if __name__ == '__main__':
    students = []
    for _ in range(int(input())):
        name = input()
        score = float(input())


        students.append([name, score])

    scores = [score for name, score in students]

    unique_scores = sorted(set(scores))


    second_lowest = unique_scores[1]
    names = [name for name, score in students if score == second_lowest]
    names.sort()
    for name in names:
        print(name)

```

```python
#Text Wrap


def wrap(string, max_width):
    result = ""
    for i in range(0, len(string), max_width):
        result += string[i:i+max_width] + "\n"
    return result.rstrip()
    return
```

```python
#Piling Up!
# Enter your code here. Read input from STDIN. Print output to STDOUT
t = int(input())

for _ in range(t):
    n = int(input())
    cube_sizes = list(map(int, input().split()))

    left_index = 0
    right_index = n - 1
    current_top = float('inf')
    stackable = True

    while left_index <= right_index:
        if cube_sizes[left_index] >= cube_sizes[right_index]:
            next_cube = cube_sizes[left_index]
            left_index += 1
        else:
            next_cube = cube_sizes[right_index]
            right_index -= 1

        if next_cube <= current_top:
            current_top = next_cube
        else:
            stackable = False
            break

    print("Yes" if stackable else "No")
```

```python
#Collections.deque()
# Enter your code here. Read input from STDIN. Print output to STDOUT
n = int(input())
dq = []

for _ in range(n):
    operation = input().split()

    if operation[0] == "append":
        dq.append(int(operation[1]))
    elif operation[0] == "appendleft":
        dq.insert(0, int(operation[1]))
    elif operation[0] == "pop":
        if dq:
            dq.pop()
    elif operation[0] == "popleft":
        if dq:
            dq.pop(0)

print(*dq)
```
